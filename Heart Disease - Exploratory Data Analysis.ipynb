# #Heart Disease EDA - Correlation - (JointPlot)


```python
sns.jointplot(x = 'Age', y = 'MaxHR', data = df)
```

```python
sns.jointplot(x = 'Age', y = 'Age', data = df)  # Perfect co-relation
```

```python
sns.jointplot(x = 'HeartDisease', y = 'Oldpeak', data = df)  # Zero co-relation
```

```python
sns.jointplot(x = 'Age', y = 'MaxHR', data = df, kind = 'hex')
```

```python
sns.jointplot(x = 'Age', y = 'MaxHR', data = df, kind ='reg')
```
